```python
import pandas as pd

# 1. Load semua dataset yang sudah bersih
df1 = pd.read_csv('nli_eng.csv', sep=';')
df2 = pd.read_csv('nli_idn.csv', sep=';')
df3 = pd.read_csv('nli_idn-ocr.csv', sep=';')

# 2. Tambahkan kolom penanda (Opsional tapi sangat disarankan)
df1['language_type'] = 'eng'
df2['language_type'] = 'idn'
df3['language_type'] = 'idn-ocr'

# 3. Gabungkan secara vertikal
df_master = pd.concat([df1, df2, df3], ignore_index=True)

# 4. Simpan hasil merge master
output_master = 'nli_all.csv'
df_master.to_csv(output_master, sep=';', index=False, encoding='utf-8')

# 5. Tampilkan Ringkasan
print("="*30)
print("SUMMARY MASTER MERGE")
print("="*30)
print(f"Total baris ENG     : {len(df1)}")
print(f"Total baris IDN     : {len(df2)}")
print(f"Total baris IDN-OCR : {len(df3)}")
print("-" * 30)
print(f"TOTAL MASTER DATA: {len(df_master)}")
print(f"File disimpan di : {output_master}")
print("="*30)

# Tampilkan beberapa baris awal
df_master.head()
```

```python
# Validasi Master vs Referensi
df_ref = pd.read_csv(r'C:\Users\Liza\Documents\Kerja Praktik FIXX\dataset\combined_dataset_all.csv', sep=';', encoding='utf-8')

for col in ['question', 'answer']:
    df_master[col] = df_master[col].astype(str).str.strip()
    df_ref[col] = df_ref[col].astype(str).str.strip()

validation = pd.merge(
    df_ref[['question', 'answer']], 
    df_master[['question', 'answer']], 
    on=['question', 'answer'], 
    how='outer', 
    indicator=True
)

print("-" * 30)
print("VALIDATION SUMMARY")
print("-" * 30)
print(f"Total Ref    : {len(df_ref)}")
print(f"Total Master : {len(df_master)}")
print(f"Match        : {len(validation[validation['_merge'] == 'both'])}")
print(f"Missing      : {len(validation[validation['_merge'] == 'left_only'])}")
print(f"Extra        : {len(validation[validation['_merge'] == 'right_only'])}")
print("-" * 30)
```
