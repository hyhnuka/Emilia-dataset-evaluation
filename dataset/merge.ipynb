## MERGE TIAP DIREKTORI

```python
import os
import json
import pandas as pd
import csv

# Langsung arahkan ke folder 'eng'
BASE_DIR = r"C:\Users\Liza\Documents\Kerja Praktik 2\dataset\eng"

data = []
total_files = 0
total_items = 0

print(f"📂 Memulai proses di folder: {BASE_DIR}")

if not os.path.exists(BASE_DIR):
    print("❌ Folder tidak ditemukan!")
else:
    for root, dirs, files in os.walk(BASE_DIR):
        for filename in files:
            if not filename.lower().endswith(".json"):
                continue

            file_path = os.path.join(root, filename)
            print(f"📄 Processing: {filename}")
            
            total_files += 1

            # ===== PARSE NAMA FILE =====
            # 1. Ambil nama lengkap tanpa .json untuk kolom source
            source_full_name = filename.replace(".json", "")
            
            # 2. Tetap split untuk menentukan 'generate_from'
            parts = source_full_name.split("_")
            if len(parts) >= 2:
                generate_from = parts[1]
                if generate_from.upper() != "GPT":
                    generate_from = "GROK"
            else:
                generate_from = "UNKNOWN"

            # ===== LOAD JSON =====
            try:
                with open(file_path, "r", encoding="utf-8") as f:
                    content = json.load(f)

                if isinstance(content, dict):
                    content = content.get("data", [])

                if not isinstance(content, list):
                    continue

                added_count = 0
                for item in content:
                    question = item.get("Question") or item.get("question") or ""
                    answer = item.get("Answer") or item.get("answer") or ""

                    if question.strip() == "" or answer.strip() == "":
                        continue

                    data.append({
                        "question": question.strip(),
                        "answer": answer.strip(),
                        "from": generate_from,
                        "source": source_full_name, # Sekarang berisi nama file lengkap (minus .json)
                        "folder": "eng"
                    })
                    added_count += 1

                total_items += added_count
                print(f"    ✅ Ditambahkan: {added_count}")

            except Exception as e:
                print(f"  ❌ ERROR pada file {filename}: {e}")

# ===== SAVE =====
df = pd.DataFrame(data)

if df.empty:
    print("\n⚠️ Data kosong!")
else:
    # 1. Tampilkan duplikat untuk pengecekan
    duplicates = df[df.duplicated(subset=["question", "answer"], keep=False)]
    if not duplicates.empty:
        print(f"\n🔍 Ditemukan {len(duplicates)} baris duplikat. Detail di: cek_duplikat_eng.csv")
        duplicates.sort_values("question").to_csv("cek_duplikat_eng.csv", index=False, sep=";")

    # 2. Bersihkan duplikat
    df_clean = df.drop_duplicates(subset=["question", "answer"])
    
    output_name = "combined_dataset_eng.csv"
    df_clean.to_csv(
        output_name,
        index=False,
        encoding="utf-8",
        sep=";",
        quoting=csv.QUOTE_ALL
    )

    print("\n" + "="*30)
    print("📊 SUMMARY (FOLDER ENG)")
    print(f"Total file JSON diproses       : {total_files}")
    print(f"Total data mentah              : {total_items}")
    print(f"Total data unik (final)        : {len(df_clean)}")
    print(f"File final disimpan di         : {output_name}")
    print("="*30)
```

## MERGE ALL

```python
import os
import json
import pandas as pd
import csv

BASE_DIR = r"C:\Users\Liza\Documents\Kerja Praktikm2\dataset"

data = []
total_files = 0
total_items = 0

for subdir in ["eng", "idn", "idn-ocr"]:  # FIX underscore
    main_folder = os.path.join(BASE_DIR, subdir)

    print(f"\n📂 Masuk folder utama: {main_folder}")

    if not os.path.exists(main_folder):
        print("❌ Folder tidak ditemukan!")
        continue

    for root, dirs, files in os.walk(main_folder):
        print(f"\n  📁 Subfolder: {root}")
        print(f"     Jumlah file: {len(files)}")

        for filename in files:
            if not filename.lower().endswith(".json"):
                continue

            file_path = os.path.join(root, filename)
            print(f"    📄 Processing: {filename}")

            total_files += 1

            # ===== PARSE NAMA FILE =====
            name = filename.replace(".json", "")
            parts = name.split("_")

            if len(parts) < 3:
                print("    ⚠️ Skip (nama tidak sesuai):", filename)
                continue

            generate_from = parts[1]
            source = "_".join(parts[2:])

            if generate_from.upper() != "GPT":
                generate_from = "GROK"

            # ===== LOAD JSON =====
            try:
                with open(file_path, "r", encoding="utf-8") as f:
                    content = json.load(f)

                if isinstance(content, dict):
                    content = content.get("data", [])

                if not isinstance(content, list):
                    print("    ❌ Bukan list")
                    continue

                added_count = 0

                for item in content:
                    question = item.get("Question") or item.get("question") or ""
                    answer = item.get("Answer") or item.get("answer") or ""

                    if question.strip() == "" or answer.strip() == "":
                        continue

                    data.append({
                        "question": question.strip(),
                        "answer": answer.strip(),
                        "from": generate_from,   # disamain sama CSV kamu
                        "source": source,
                        "folder": subdir
                    })

                    added_count += 1

                total_items += added_count
                print(f"    ✅ Ditambahkan: {added_count}")

            except Exception as e:
                print(f"    ❌ ERROR: {e}")

# ===== SAVE =====
df = pd.DataFrame(data)

if len(df) == 0:
    print("\n⚠️ Data kosong! Cek lagi struktur JSON.")
else:
    df = df.drop_duplicates(subset=["question", "answer"])
    df = df[df["question"] != ""]
    df = df[df["answer"] != ""]

    df.to_csv(
        "combined_dataset.csv",
        index=False,
        encoding="utf-8",
        sep=";",                 # 🔥 penting
        quoting=csv.QUOTE_ALL    # 🔥 biar aman dari koma & newline
    )

    print("\n======================")
    print("📊 SUMMARY")
    print("Total file diproses:", total_files)
    print("Total data masuk:", len(df))
    print("======================")
```
